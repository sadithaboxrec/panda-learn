# Lambda and Custom Functions

```python
import pandas as pd
bios = pd.read_csv('../data/bios.csv')
```

```python
bios['height_category']=bios['height_cm'].apply(lambda x : 'short' if x<165.0 else ('Average' if x<185.0 else 'tall') )
```

```python
bios.head()
```

```python
def categorize_athlete(row):
    if row['height_cm']<175 and row['weight_kg']<70:
        return 'Lightweight'
    elif row['height_cm']<185 or row['weight_kg']<80:
        return 'Middleweight'
    else:
        return 'Heavyweight'

bios['Category'] = bios.apply(categorize_athlete, axis=1)
# 1 is for rows and 0 is dor columns. we specify that this is focused on rows
```
