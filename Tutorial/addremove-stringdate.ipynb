```python
import pandas as pd
bios = pd.read_csv('../data/bios.csv')
```

```python
bios.head()
```

```python
bios_new = bios.copy()
```

```python
bios_new['first_name'] = bios_new['name'].str.split(' ').str[0]
bios_new
```

```python
bios_new.query('first_name == "Keith" ')
```

```python
bios_new.info()
```

```python
bios_new['born_datetime']= pd.to_datetime(bios_new['born_date'])

# converting to datetime object
```

```python
bios_new
```

```python
bios_new.info()
```

```python
bios_new['born_year'] = bios_new['born_datetime'].dt.year

# dt ; like .str . give access to lots of things

bios_new[['name', 'born_year']]
```

```python
import calendar
bios_new['born_leap_year'] = bios_new['born_datetime'].dt.year.apply(calendar.isleap)
bios_new[['name', 'born_leap_year']]
```

```python
# Filter only the rows where born year is a leap year
leap_year_born = bios_new[bios_new['born_datetime'].dt.year.apply(calendar.isleap)]

leap_year_born[['name', 'born_datetime']]
```

```python
# saving newly created data

bios_new.to_csv('../data/bios_new.csv', index=False)
```
