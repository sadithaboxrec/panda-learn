# Adding Removing Columns

```python
import pandas as pd
coffee=pd.read_csv('../warmup-data/coffee.csv')

coffee.head()
```

```python
coffee['price']=4.99
coffee.head()
```

```python
# one price for espresso and one price for latte

import numpy as np
coffee['new_price']=np.where(  coffee ['Coffee Type']=='Espresso',3.99,5.99  )
coffee.head()
```

```python
coffee
```

```python
coffee.drop(0)
# index
```

```python
# in previous we didn't permananetly modified the actual data frame.
# we still have 0
coffee
```

```python
coffee.drop(columns=['price'])
```

```python
# with drop we still didn't permanently modified it
coffee
```

```python
coffee.drop(columns=['price'],inplace=True)

# run once, permanently modified
```
