# Data Accessing loc,iloc

```
#coffee.loc[rows,columns]
coffee.loc[0]

#locals allow us to filter by rows and columns
```

```
coffee.loc[[1,2,5] ]
```

```
coffee.loc[1:4]
```

```
coffee.loc[1:4,"Day"]
```

```
coffee.loc[1:4, ["Day", "Units Sold"] ]

#  rows with labels 1 through 4 (inclusive).
```

```
# All rows with certain columns
coffee.loc[:,["Day", "Units Sold"]]
```

```
# iloc use index
coffee.iloc[:,[0,2]]
```

```
coffee.iloc[0:5,[0,2]]

#rows 0 through 4 (Python slicing is exclusive on the end).
```

# Sort values

```
coffee.sort_values("Units Sold")
# increasing order
```

```
coffee.sort_values("Units Sold",ascending=False)
#decreasing orders
```

```
coffee.sort_values(["Units Sold","Coffee Type"],ascending=False)
```

```
coffee.sort_values(["Units Sold","Coffee Type"],ascending=[0,1])

# for units sold in not ascending but Coffee type is ascending order
```
