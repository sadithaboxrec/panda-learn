### To have multiple condition

```python
bios[ (bios['height_cm']>215) & (bios['born_country']=='USA'  )]
```

### filter based on string operation

```python
bios[bios['name'].str.contains('Keith') ]
```

```python
bios[bios['name'].str.contains('keith',case=False) ]
```

```python
bios[bios['name'].str.contains('keith|patrick',case=False) ]

# | or operator
```

```python
repeated_letters = bios[bios['name'].str.contains(r'(.)\1', na=False)]
repeated_letters

# (.)\1: Match any character (.) followed by the same character \1.
# So it matches repeated letters, like:
```

```python
bios[bios['name'].str.contains('keith|patrick',case=False ,regex=False ) ]

# disabling regex expressions
```

```python
bios[bios['born_country'].isin(['USA','FRA','GBR']) ]
```

```python
bios[bios['born_country'].isin(['USA','FRA','GBR']) & (bios['name']).str.startswith('Keith') ]
```

```python
bios.query('born_country=="USA"')
```

```python
bios.query('born_country=="USA" and born_city=="Newark" ')
```

```python
bios.query('born_country in ["USA", "FRA", "GBR"] and name.str.contains("Keith")')
```
